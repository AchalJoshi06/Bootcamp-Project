# Group by & Aggregating

```python
import pandas as pd
```

```python
df=pd.read_csv(r"D:\DARK 's PYTHON PROGRAM\JUPITER NOTEBOOK\Flavors.csv")
```

```python
df
```

```python
group_by_frame = df.groupby('Base Flavor')
```

```python
group_by_frame.mean(numeric_only=True)
```

```python
group_by_frame.count()
```

```python
group_by_frame.min(numeric_only=True)
```

```python
# group_by_frame.max(numeric_only=True)
# df.groupby('Base Flavor').sum(numeric_only=True)

df.groupby('Base Flavor').agg({'Flavor Rating':['mean','min','max','sum'],'Texture Rating' : ['mean','min','max','sum']})
```

```python
df
```

```python
group_by_frames= df.groupby(['Base Flavor','Liked'])
```

```python
group_by_frames.mean(numeric_only=True)
```

```python
df.groupby(['Base Flavor','Liked']).agg({'Flavor Rating':['mean','min','max','sum'],'Texture Rating' : ['mean','min','max','sum']})
```
