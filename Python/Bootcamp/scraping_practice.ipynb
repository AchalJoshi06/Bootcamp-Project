```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url= 'https://webscraper.io/test-sites/tables'
page = requests.get(url)
soup = BeautifulSoup(page.text,'html')
```

```python
print(soup)
```

```python
table=soup.find('table',class_="table table-bordered")
```

```python
scrap_table=table.find_all('th')
```

```python
scrap_table_titles = [title.text.strip() for title in scrap_table]
scrap_table_titles
```

```python
df = pd.DataFrame(columns=scrap_table_titles)
df
```

```python
column_data=table.find_all('tr')
for row in column_data[1:]:
    row_data=row.find_all('td')

    individual_row_table = [data.text.strip() for data in row_data]


    length=len(df)
    df.loc[length] = individual_row_table
```

```python
df.set_index('#')
```
